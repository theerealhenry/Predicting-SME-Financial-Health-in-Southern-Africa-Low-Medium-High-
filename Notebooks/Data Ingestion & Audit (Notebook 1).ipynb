```python
# Sample values (quick look at each column’s typical categories)

def sample_values(df, col, n=10):
    vals = df[col].dropna().astype(str).unique()
    return list(vals[:n])

samples = []
for col in train.columns:
    samples.append({
        "columns": col,
        "dtype": str(train[col].dtype),
        "n_unique": train[col].nunique(dropna=True),
        "sample-values": sample_values(train, col, n=10)
    })

sample_df = pd.DataFrame(samples).sort_values(["dtype", "n_unique"],  ascending=[True, False])
sample_df.head(40)
```

```python
# Data quality check 1: ID uniqueness

# check if IDs are unique in train and test
train_unique = train["ID"].is_unique
test_unique = test["ID"].is_unique

print("Train IDs unique?", train_unique)
print("Test IDs unique?", test_unique)

# if not unique, show duplicates
if not train_unique:
    dup_ids = train[train["ID"].duplicated()]["ID"].value_counts()
    print("Duplicated IDs in train:\n", dup_ids.head(20))

if not test_unique:
    dup_ids = test[test["ID"].duplicated()]["ID"].value_counts()
    print("Duplicated IDs in test:\n", dup_ids.head(20))
```

```python
# Data quality check 2: train-test schema match
# We expect: train.columns = test.columns + Target

train_cols = set(train.columns)
test_cols = set(test.columns)

only_in_train = sorted(list(train_cols - test_cols))
only_in_test = sorted(list(test_cols - train_cols))

print("Columns only in train:", only_in_train)
print("Columns only in test:", only_in_test)
```

```python
# Data quality check 3: Categorical string normalization

# Identify categorical columns (object type)
from pandas.api.types import is_string_dtype, is_object_dtype, is_categorical_dtype

cat_cols = [
    c for c in train.columns
    if c != TARGET_COL and (
        is_string_dtype(train[c]) or 
        is_object_dtype(train[c]) or 
        is_categorical_dtype(train[c])
    )
]

print("Number of categorical columns:", len(cat_cols))
print("First few categorical columns:", cat_cols[:10])


def clean_text(x):
    if pd.isna(x):
        return x
    x = str(x).strip()

    # Fix common broken apostrophe encoding
    x = x.replace("â€™", "'").replace("’", "'").replace("`", "'")

    # Standardize spacing
    x = " ".join(x.split())

    # Lowercase for consistency
    x_low = x.lower()

    # Normalize common "don't know" variants
    dont_know_variants = {
        "dont know", "don't know", "do not know", "don’t know", "donâ€™t know"
    }
    if x_low in dont_know_variants:
        return "Don't know"

    # Normalize yes/no (optional but helpful)
    if x_low in {"yes", "y"}:
        return "Yes"
    if x_low in {"no", "n"}:
        return "No"

    # Title-case common multi-word categories optionally:
    # Keep original for non-trivial categories (e.g., "Used to have but don't have now")
    # But we can fix apostrophe form:
    return x.replace("Don't", "Don't")

# Apply cleaning to train and test
train_clean = train.copy()
test_clean = test.copy()

for c in cat_cols:
    train_clean[c] = train_clean[c].apply(clean_text)
    test_clean[c] = test_clean[c].apply(clean_text)

# Check if we reduced weird variants in a suspicious column (example)
# Replace with a real column you see issues in
if "attitude_worried_shutdown" in train.columns:
    example_col = "attitude_worried_shutdown"
elif len(cat_cols) > 0:
    example_col = cat_cols[0]
else:
    example_col = None

print("example_col:", example_col)
```

```python
# Data quality check 4: Numeric coercion (turn text money into numbers)

# Candidate numeric columns based on your variable definitions
numeric_candidates = [
    "owner_age",
    "personal_income",
    "business_expenses",
    "business_turnover",
    "business_age_years",
    "business_age_months"
]

numeric_candidates = [c for c in numeric_candidates if c in train.columns]

def to_numeric_safe(s):
    # Convert to string, remove commas, then numeric
    return pd.to_numeric(
        s.astype(str).str.replace(",", "", regex=False),
        errors="coerce"
    )

train_num = train_clean.copy()
test_num  = test_clean.copy()

for c in numeric_candidates:
    # Before type
    before_dtype = train_num[c].dtype

    train_num[c] = to_numeric_safe(train_num[c])
    test_num[c]  = to_numeric_safe(test_num[c])

    after_dtype = train_num[c].dtype
    print(f"{c}: {before_dtype} -> {after_dtype} | missing now: {train_num[c].isna().mean():.2%}")
```

```python
# Data quality check 5: Outliers (money columns & extreme values)

outlier_report = []

for c in ["personal_income", "business_expenses", "business_turnover"]:
    if c in train_num.columns:
        s = train_num[c].dropna()
        if len(s) > 0:
            outlier_report.append({
                "column": c,
                "min": s.min(),
                "p01": s.quantile(0.01),
                "p50": s.quantile(0.50),
                "p99": s.quantile(0.99),
                "max": s.max()
            })

pd.DataFrame(outlier_report)
```

```python
# Create the final audit tables (Deliverables)

audit_rows = []
for col in train_num.columns:
    audit_rows.append({
        "column": col,
        "dtype": str(train_num[col].dtype),
        "missing_pct": round(train_num[col].isna().mean() * 100, 2),
        "n_unique": train_num[col].nunique(dropna=True),
        "sample_values": sample_values(train_num, col, n=5)
    })

audit_table = pd.DataFrame(audit_rows).sort_values(["missing_pct", "n_unique"], ascending=[False, False])
audit_table.head(50)
```

```python
# Save cleaned datasets for modeling

train_num.to_csv("train_cleaned.csv", index=False)
test_num.to_csv("test_cleaned.csv", index=False)

print("Saved: train_cleaned.csv and test_cleaned.csv")
```
